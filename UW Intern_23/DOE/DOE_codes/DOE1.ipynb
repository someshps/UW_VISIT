```python
from pyDOE2 import ccdesign

# Define the number of levels for each factor
quantitative_levels = 3  # Number of levels for quantitative factors

# Convert quantitative_levels to integer
quantitative_levels = int(quantitative_levels)

# Generate the CCD design matrix
design = ccdesign(2, center=[quantitative_levels/2, quantitative_levels/2], alpha='orthogonal')

# Scale the design for quantitative factors
scaled_design = design.copy()
scaled_design[:, :2] = scaled_design[:, :2] * (quantitative_levels - 1)

# Define the levels for the qualitative factor
qualitative_levels = ['Low', 'High']

# Assign qualitative factor levels to the design
design_with_qualitative = scaled_design.copy()
design_with_qualitative[:, 2] = qualitative_levels

# Print the design matrix
print("CCD Design Matrix:")
print(design_with_qualitative)
```

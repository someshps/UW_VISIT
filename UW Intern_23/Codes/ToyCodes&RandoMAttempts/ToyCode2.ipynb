```python
import numpy as np
from skopt import gp_minimize
from skopt.space import Real
from skopt.utils import use_named_args
```

```python
# Given points
xs = [0, 0.52, 0.79, 1.04, 1.5]
ys = [0, 0.5, 0.707, 0.866, 1]
zs = [0, 0.5, 0.707, 0.866, 1]
arr_xs = np.array(xs)
arr_ys = np.array(ys)
arr_zs = np.array(zs)
```

```python
# Define the objective function
def objective_function(x):
    # Calculate the distance between the given points and the current point
    distances = np.sqrt((x[0] - arr_xs)**2 + (x[1] - arr_ys)**2)
    # Find the closest point
    closest_index = np.argmin(distances)
    # Return the value of the closest point
    return arr_zs[-closest_index]
```

```python
space = [Real(0, 3.14), Real(0, 1)]
```

```python
result = gp_minimize(
    objective_function,
    space,
    acq_func="EI",  # Thomson sampling
    n_calls=20,     # Number of iterations
    random_state=42
)
```

```python
best_x = result.x
best_y = result.fun
print("Optimized function value:", best_y)
print("Optimized parameters:", best_x)
```

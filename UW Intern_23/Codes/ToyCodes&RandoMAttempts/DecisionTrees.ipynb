```python
from sklearn import tree
```

```python
X = [[2, 5, 13], [20, 15, 6.5], [2, 5, 6.5], [2, 15, 13], [10, 5, 13], [10, 15, 6.5], [10, 5, 6.5], [10, 15, 13], [20, 5, 13], [2, 15, 6.5], [20, 5, 6.5], [20, 15, 13], [0, 0 ,0]]
```

```python
Y = ['F','F','F','P','P','F','P','F','F','P','P','P','F']
```

```python
clf = tree.DecisionTreeClassifier(max_depth = 10)
clf = clf.fit(X, Y)
```

```python
tree.plot_tree(clf)
```

```python
import graphviz
```

```python
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data)
graph.render("better tree") 
graph
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, 
                    feature_names=['Passes','Speed','SOD'],  
                   class_names=['P', 'F'],  
                     filled=True, rounded=True,  
                    special_characters=True)  
```

```python
graph = graphviz.Source(dot_data)
```

This analysis is based on intuition that 

Trying for the worst case of
1. Speed (15) and SOD (13): Number of passes for which it passes = 16;
2. Number of Passes (1) and Speed (15): Fails for both SOD values;
3. SOD (13) and Number of Passes (1): Test passes for no value of speed;

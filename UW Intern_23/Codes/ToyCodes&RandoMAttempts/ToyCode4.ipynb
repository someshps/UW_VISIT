Linear Regression followed by gradient descent

```python
import numpy as np
```

```python
# Define the input attributes (x) and output (y) data
x = np.array([[1, 2, 3, 4],
              [2, 4, 6, 8],
              [3, 6, 9, 12],
              [4, 8, 12, 16]])
y = np.array([10, 20, 30, 40])
```

```python
# Add a column of ones to x for the intercept term
intercept = np.ones((x.shape[0], 1))
x = np.concatenate((intercept, x), axis=1)
```

```python
# Initialize the coefficients (theta) randomly
np.random.seed(42)
theta = np.random.rand(x.shape[1])
```

```python
# Define the learning rate and number of iterations
learning_rate = 0.01
num_iterations = 1000
```

```python
# Perform gradient descent
for i in range(num_iterations):
    # Calculate the predicted values
    y_pred = np.dot(x, theta)
    
    # Calculate the gradient
    gradient = np.dot(x.T, (y_pred - y))
    
    # Update the coefficients
    theta -= learning_rate * gradient
```

```python
# Get the optimal values of the input attributes
optimal_x = x[np.argmax(y_pred), 1:]

# Print the coefficients and the optimal input attributes
print("Coefficients:", theta[1:])
print("Optimal input attributes:", optimal_x)
```

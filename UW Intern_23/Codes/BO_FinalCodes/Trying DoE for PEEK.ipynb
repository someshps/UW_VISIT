```python
from SALib.sample import sobol_sequence

# Set the number of dimensions and number of points to generate
num_dimensions = 4
num_points = 12

# Set the ranges for each dimension
ranges = [
    [1, 10],     # Range for the first dimension (1 to 10)
    [15, 45],    # Range for the second dimension (15 to 45)
    [2, 8],      # Range for the third dimension (2 to 8)
    [100, 120]   # Range for the fourth dimension (100 to 120)
]

# Generate Sobol sequence
sobol_sequence = sobol_sequence.sample(num_dimensions, num_points)

# Scale the Sobol sequence to the desired ranges
scaled_sequence = []
for i in range(num_dimensions):
    scaled_sequence.append([sobol_sequence[:, i] * (ranges[i][1] - ranges[i][0]) + ranges[i][0]])

# Transpose the scaled sequence to get the desired format
sobol_points = list(map(list, zip(*scaled_sequence)))

# Print the generated Sobol points
for point in sobol_points:
    print(point)

```
